# Week 01 — Random numbers, quartiles and plots

This notebook generates a numpy array of random numbers, computes quartiles, and shows a histogram and a boxplot. It also prints basic system information.

```python
# Create a numpy array with random numbers
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=42)  # deterministic seed for reproducibility
data = rng.normal(loc=0, scale=1, size=1000)

# Show a short preview
print('Data preview (first 10 values):', data[:10])
```

```python
# Calculate quartiles (0th, 25th, 50th, 75th, 100th percentiles)
quartiles = np.percentile(data, [0, 25, 50, 75, 100])
print('Quartiles (min, Q1, median, Q3, max):', quartiles)
```

```python
# Histogram
%matplotlib inline
plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color='steelblue', edgecolor='black')
plt.title('Histogram of random data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 2))
plt.boxplot(data, vert=False)
plt.title('Boxplot of random data')
plt.xlabel('Value')
plt.show()
```
